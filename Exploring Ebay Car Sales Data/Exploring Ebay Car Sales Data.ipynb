## Part VI - Dealing with Incorrect Registration Year Data

- Realistically, we would check the year the ad was created and make it so that the registration year is not higher than that, however in this case, because all the ads were created in 2015-2016, 2016 will be the upper limit of the registration years included
- After some research, it appears that the earliest registration date was 1903 so that will be ther lower limit of the registration years included

```python
# Calculate distribution of values
autos['registration_year'].value_counts(normalize=True)
```

```python
# Percentage of car listings that fall outside the 1900 to 2016
# ~ = NOT
(~autos['registration_year'].between(1903,2016)).sum()/autos.shape[0]*100
```

```python
# delete values that are not between 1903 and 2016
autos=autos[autos['registration_year'].between(1903,2016)]
autos.shape
```

```python
# Recalculate distribution of remaining values
autos['registration_year'].value_counts(normalize=True)
```

- frequencies increased for all remaining values because the there are less values involved
- most values are 1985 and above

## Part VII - Exploring Price by Brand

```python
all_brands=autos['brand'].value_counts(normalize=True)[:20]
print(all_brands)
print(all_brands.shape[0])
print(all_brands.sum())
print(all_brands.index)
```

```python
above_this=autos['brand'].value_counts(normalize=True)
above_this=above_this[above_this>0.05]
print(above_this)
print(above_this.shape[0])
print(above_this.sum())
print(above_this.index)
```

- The brands that make up the top 20 brands come from a varieties of countries and classes
- The top 20 brands make up almost 93% of the brands while the top 5 make up about 68% of the brands. Because the top 20 covers a significantly higher percentage of the listings

```python
aggdat={}
for brand in all_brands.index:
    bmean=autos['price'][autos['brand'] == brand].mean()
    aggdat[brand]=bmean
print(aggdat)
#sort dictionary by values https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value
```

- Most brand are within a range of \$2,000 to \$10,000 while others are more expensive and around \$30,000 - \$40,000

## Part VIII - Storing Aggregate Data in a DataFrame

```python
miledat={}
for brand in all_brands.index:
    mmean=autos['odometer_km'][autos['brand'] == brand].mean()
    miledat[brand]=mmean
print(miledat)
```

```python
pd_aggdat=pd.Series(aggdat)
pd_miledat=pd.Series(miledat)
df_dat=pd.DataFrame(pd_aggdat, columns=['mean_price'])
df_dat['mean_milage']=pd_miledat
df_dat
```

```python
df_dat.sort_values(by='mean_price')
```

```python
df_dat.sort_values(by='mean_milage')
```
