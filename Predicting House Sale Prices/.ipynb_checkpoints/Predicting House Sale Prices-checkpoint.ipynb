# Predicting House Sale Prices Using Linear Regression
Going to use linear regression model with selected features to predict sale prices for houses in Ames, Iowa from 2006 to 2010. The dataset will be broken up into a training and testing set and the performance will be evaluated based on the root mean square error. In order to quickly iterate on different models a pipline of functions will be implemented.

- Import libraries and classes into the environment

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
from sklearn.model_selection import cross_val_score, KFold
```

```python
ogdf = pd.read_csv('AmesHousing.tsv', delimiter ='\t')
ogdf
```

```python
def transform_features(df):
    return df
```

```python
def select_features(df):
    return df[['Gr Liv Area','SalePrice']]
```

```python
def train_and_test(df):
    train = df[:1460]
    test = df[1460:]
    
    model_train = train
    numerical_model_train = model_train.drop('SalePrice', axis=1)\
        .select_dtypes(include=['float','int'])
    
    model_test = test
    numerical_model_test = model_test.drop('SalePrice', axis=1)\
        .select_dtypes(include=['float','int'])
    
    lr = LinearRegression()
    lr.fit(numerical_model_train, model_train['SalePrice'])
    predictions_test = lr.predict(numerical_model_test)
    rmse = (mean_squared_error(model_test['SalePrice'], predictions_test))**0.5
    
    return rmse
```

### Redefine The transform_features() Function By Implementing the Temporary WorkFlow Steps Into It

```python
def transform_features(df):
    dfc = df.copy() #1
    above5 = dfc.columns[dfc.isnull().mean()>=0.05]#3
    dfc = dfc.drop(columns = above5)#3
    #everything from here has less than 5% missing values
    numeric = dfc.select_dtypes(include=['int','float']).columns#4
    modes = pd.Series(dfc[numeric].mode().iloc[0])#5
    dfc[numeric] = dfc[numeric].fillna(modes)#5
    texts = dfc.select_dtypes(include=['object']).columns #6
    textsmissing = dfc[texts].columns[dfc[texts].isnull().sum()>0] #6
    dfc = dfc.drop(columns = textsmissing) #6
    dfc['Yrs Until Sold'] = dfc['Yr Sold'] - dfc['Year Built']#7
    dfc['Yrs After Remod/Add'] = dfc['Yr Sold'] - dfc['Year Remod/Add']#7
    droprows1 = dfc[dfc['Yrs Until Sold']<0].index#7
    droprows2 = dfc[dfc['Yrs After Remod/Add']<0].index#7
    droprows = droprows1.append(droprows2)#7
    dfc = dfc.drop(index = droprows)#7
    notneeded = ['Year Built','Year Remod/Add']#7
    dfc = dfc.drop(columns = notneeded)#7
    identifiers = ['PID','Order'] #8
    dfc = dfc.drop(columns = identifiers) #8
    leaks = ['Mo Sold','Yr Sold','Sale Type','Sale Condition'] #9
    dfc = dfc.drop(columns = leaks) #9
    return dfc
```

**Update the logic for the select_features() function**
- This function will only execute the features included in the key and will use the same threshold values to check RMSE values
- This function should take in the new, modified train and test data frames that were returned from transform_features()
- The function has to be more general because the transform_features() function will not always spit out the same features

1) Delete numerical columns that have an absolute correlation coefficient value with SalePrice lower than the threshold
    - Identify numerical columns from the result of the transform_feature()
    - Index and drop columns that do not have a high enough correlation coefficient
2) Identify all the potential text columns in the original dataframe that are meant to be categorical
    - Determine if that column is within the output of the transform_feature()
    - Count the number of unique categories in each column and drop those that are above the unique number of categories threshold
3) Convert the text columns that meet the criteria to the categorical data type and create dummy variables for them
    - The only numerical column that is meant to categorical is MS SubClass and this will has a low absolute correlation coefficient (<0.01), so it most likely will be knocked out in step 1

Note: Categorical columns are nominal and not ordinal, meaning that the values cannot be compared, but they can be differentiated

ordinal_categories = ['Lot Shape','Utilities','Land Slope','Exter Qual','Exter Cond','Bsmt Qual','Bsmt Cond','Bsmt Exposure','BsmtFin Type 1','BsmtFin Type 2','Heating QC','Electrical','Kitchen Qual','Functional','Fireplace Qu','Garage Finish','Garage Qual','Garage Cond','Paved Drive','Pool QC','Fence']

```python
def select_features(df,abs_corr_coef=0.4,n_uniq_ctg=10):
    
    dfs = df.copy()
    numeric = dfs.select_dtypes(include=['int','float']).columns#1
    corr_SalePrice = dfs[numeric].corr()['SalePrice']#1
    col_below_coef = corr_SalePrice[abs(corr_SalePrice)<abs_corr_coef]#1
    dfs = dfs.drop(columns = col_below_coef.index)#1
    
    potn_categ = ['PID','MS SubClass','Ms Zoning','Street','Alley',
                 'Land Contour','Lot Config','Neighborhood','Condition1',
                 'Condition2','BldgType','House Style','Roof Style',
                 'Roof Matl','Exterior 1st','Exterior 2nd','Mas Vnr Type',
                 'Foundation','Heating','Central Air','Garage Type',
                 'Misc Feature','Sale Type','Sale Condition']#2
    current_cols = dfs.columns#2
    current_categ = []
    for col in current_cols:#2
        if col in potn_categ:#2
            current_categ.append(col)#2
    categ_uniq = dfs[current_categ].apply(lambda x:len(x.unique()))#2
    col_uniq = categ_uniq[categ_uniq > n_uniq_ctg].index#2
    dfs = dfs.drop(columns = col_uniq)#2
    
    text = dfs.select_dtypes(include = ['object']).columns#3
    for col in text:#3
        dfs[col] = dfs[col].astype('category')#3
    dfs = pd.concat([dfs, pd.get_dummies(dfs.select_dtypes(include=['category']))
                    ],axis=1).drop(columns = text)
    return dfs
```

## Part IV - Train and Test
When iterating on different features, using simple validation is a good idea. Add a parameter named k that controls the type of cross validation that occurs

```python
def train_and_test(df,k=0):
    if k==0:
        num_cols = df.select_dtypes(include=['integer','float']).columns.drop('SalePrice')
        train = df[:1460]
        test = df[1460:]
        
        lr = LinearRegression()
        lr.fit(train[num_cols],train['SalePrice'])
        te_pred = lr.predict(test[num_cols])
        rmse = (mean_squared_error(test['SalePrice'],te_pred))**0.5
        
    elif k==1:
        num_cols = df.select_dtypes(include=['integer','float']).columns.drop('SalePrice')
        shuffled_index = np.random.permutation(df.index)
        dfs = df.copy().reindex(shuffled_index)

        fold_one = dfs[:1460]
        fold_two = dfs[1460:]

        lr = LinearRegression()
        lr.fit(fold_one[num_cols],fold_one['SalePrice'])
        two_pred = lr.predict(fold_two[num_cols])
        rmse_fold_two = (mean_squared_error(fold_two['SalePrice'],two_pred))**0.5

        lr.fit(fold_two[num_cols],fold_two['SalePrice'])
        one_pred = lr.predict(fold_one[num_cols])
        rmse_fold_one = (mean_squared_error(fold_one['SalePrice'],one_pred))**0.5

        rmse = np.mean([rmse_fold_two,rmse_fold_one])
    
    else:
        df_num = df.select_dtypes(include=['integer','float'])
        train = df_num.drop('SalePrice', axis = 1)
        test = df_num['SalePrice']
        kf = KFold(k, shuffle=True, random_state=1)
        lr = LinearRegression()
        mses = cross_val_score(lr, train,test, \
                               scoring="neg_mean_squared_error", cv=kf)
        rmses = np.sqrt(np.absolute(mses))
        rmse = np.mean(rmses)
        print(rmses)
    return rmse
```

    - The reason why 'int' and 'integer' worked before is because there were no dummy columns before. Before the dummy columns were created the only numeric type were float64 and int64
    - The dtype of dummy columns are uint8, which are not picked up by using 'int' in select_dtypes, but are picked up with the use of 'integer'
    - To visualize this print: ogdf.info(), df.dtypes, df.select_dtypes(include='int').dtypes and df.select_dtypes(include='integer').dtypes

**k=1 NOTE:** 
- When shuffling the index make sure to use the **df.index** because if instead len(df) or df.shape[0] is used and that index does not exist NaN values will be filled for the new indexes
- Although it is instructed to shuffle the ordering between the rows, the key does not use the shuffled index and uses the original one, when this code uses the original one, like the key, the rmse value is the same
